```python
import pandas as pd
import math
```

```python
df_resultat_20point = pd.read_csv('new_algo_resultat_pour20points.csv', encoding='latin')
```

```python
df_resultat_20point["temps_mis_execution"] = df_resultat_20point["temps_mis_execution"] / 1e9
df_resultat_20point.columns =['nb_chemin', 'nombre_de_generation', 'distance', 'temps_mis_execution_en_second']
```

```python
min = df_resultat_20point["temps_mis_execution_en_second"].sum()/60
print("Temps total:",math.floor(min),"min",math.floor((min % 1) * 60),"sec")
```

```python
df_resultat_20point_moyenne = df_resultat_20point.groupby(["nb_chemin","nombre_de_generation"])[["distance","temps_mis_execution_en_second"]].mean().reset_index()
```

```python

df_resultat_20point_moyenne.sort_values("distance")
```
